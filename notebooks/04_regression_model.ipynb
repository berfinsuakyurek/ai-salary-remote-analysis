```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

df = pd.read_csv("../data/raw/global_ai_jobs.csv")

df["log_salary"] = np.log(df["salary_usd"])
```

```python
model = smf.ols(
    "log_salary ~ C(work_mode) + C(experience_level) + C(country)",
    data=df
).fit()
```

```python
model.summary()
```
