```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/global_ai_jobs.csv")
```

```python
import matplotlib.pyplot as plt

plt.hist(df["salary_usd"], bins=30)
plt.title("Original Salary Distribution")
plt.show()
```

```python
df["log_salary"] = np.log(df["salary_usd"])
```

```python
plt.hist(df["log_salary"], bins=30)
plt.title("Log Salary Distribution")
plt.show()
```

```python
Q1 = df["salary_usd"].quantile(0.25)
Q3 = df["salary_usd"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df_clean = df[(df["salary_usd"] >= lower) & (df["salary_usd"] <= upper)]
```

```python
print("Original:", df.shape)
print("Cleaned:", df_clean.shape)
```
