```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(df["salary_usd"], bins=40)
plt.title("Salary Distribution")
plt.xlabel("Salary (USD)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.savefig("../outputs/figures/salary_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df.boxplot(column="salary_usd", by="experience_level")
plt.title("Salary by Experience Level")
plt.suptitle("")
plt.xlabel("Experience Level")
plt.ylabel("Salary (USD)")
plt.tight_layout()
plt.savefig("../outputs/figures/salary_by_experience.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df.boxplot(column="log_salary", by="work_mode")
plt.title("Log Salary by Work Mode")
plt.suptitle("")
plt.xlabel("Work Mode")
plt.ylabel("Log Salary")
plt.tight_layout()
plt.savefig("../outputs/figures/log_salary_by_work_mode.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
top_countries = (
    df.groupby("country")["salary_usd"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .sort_values()
)

plt.figure(figsize=(8, 6))
top_countries.plot(kind="barh")
plt.title("Top 10 Countries by Average Salary")
plt.xlabel("Average Salary (USD)")
plt.ylabel("Country")
plt.tight_layout()
plt.savefig("../outputs/figures/top_countries_avg_salary.png", dpi=300, bbox_inches="tight")
plt.show()
```
