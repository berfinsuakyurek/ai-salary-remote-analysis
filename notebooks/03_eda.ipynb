```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/global_ai_jobs.csv")


df["log_salary"] = np.log(df["salary_usd"])
```

```python
df.groupby("experience_level")["salary_usd"].mean()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))

df.boxplot(column="salary_usd", by="experience_level")

plt.title("Salary by Experience Level")
plt.suptitle("")
plt.xlabel("Experience Level")
plt.ylabel("Salary USD")

plt.show()
```

```python
df.groupby("work_mode")["log_salary"].mean()
```

```python
plt.figure(figsize=(8,5))

df.boxplot(column="log_salary", by="work_mode")

plt.title("Log Salary by Work Mode")
plt.suptitle("")
plt.xlabel("Work Mode")
plt.ylabel("Log Salary")

plt.show()
```
